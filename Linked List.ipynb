## Linked List (MAM)

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
       
    def get_Next(self):
        return self.next
   
    def get_Data(self):
        return self.data
   
class Linked_List:
    def __init__(self):
        self.head = None
       
    def insertion_beg(self,d):
        new_node = Node(d)
        new_node.next = self.head
        self.head = new_node
    
    

    def traverse(self):
        current = self.head
        count = 0
        print("Linked List:")
        while current != None:
            print(current.data)
            count += 1
            current = current.get_Next()
        print("\nLength:",count)        
       
llist = Linked_List()
llist.head = Node(5)
n2 = Node(6)
n3 = Node(7)
llist.head.next = n2
n2.next = n3
llist.insertion_beg(2)
llist.traverse()    

```

### Reversing nodes of a single linked list

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
class ll2:
    def __init__(self):
        self.start=None
    def ins_end(self):
        print("Enter no of elements:")
        for i in range(int(input())):
            print("Enter element")
            n=node(int(input()))
            n.next=None
            if self.start==None:
                self.start=n
            else:
                ptr=self.start
                while(ptr.next!=None):
                    ptr=ptr.next
                ptr.next=n
    
    def length(self):
        c=0
        q=self.start
        while(q!=None):
            c+=1
            q=q.next
        return c
    
    def reverse(self):
        pn=None
        cn=self.start
        while(cn!=None):
            nn=cn.next
            cn.next=pn
            pn=cn
            cn=nn
        rev=pn
        while(rev.next!=None):
            print(rev.data)
            rev=rev.next
    def traverse(self):
        current=self.start
        while(current!=None):
            print(current.data)
            current=current.next
p=ll2()
p.ins_end()
p.reverse()
p.traverse()
```

### Reversing Nodes of LL (Recursion)

### Reverse k elements of a list

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
class ll2:
    def __init__(self):
        self.start=None
    def ins_end(self):
        print("Enter no of elements:")
        for i in range(int(input())):
            print("Enter element")
            n=node(int(input()))
            n.next=None
            if self.start==None:
                self.start=n
            else:
                ptr=self.start
                while(ptr.next!=None):
                    ptr=ptr.next
                ptr.next=n
def traverse(start):
        current=start
        while(current!=None):
            print(current.data)
            current=current.next
def kreverse(start,k):
    if start==None or k<=1:
        return start
    elif start.next==None:
        return start
    prev=start
    curr=start.next
    i=1
    prev.next=None
    while curr and i<k:
        next=curr.next
        curr.next=prev
        prev=curr
        curr=next
        i+=1
    start.next=kreverse(curr,k)
    return prev

p=ll2()
p.ins_end()
a=kreverse(p.start,3)
traverse(a)
```

### Merge 2 sorted Linked List

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
class ll2:
    def __init__(self):
        self.start=None
    def ins_end(self):
        print("Enter no of elements:")
        for i in range(int(input())):
            print("Enter element")
            n=node(int(input()))
            n.next=None
            if self.start==None:
                self.start=n
            else:
                ptr=self.start
                while(ptr.next!=None):
                    ptr=ptr.next
                ptr.next=n
    
def traverse(start):
        current=start
        while(current!=None):
            print(current.data)
            current=current.next
def merge(h1,h2):
    if h1==None:
        return h2
    if h2==None:
        return h2
    if h1.data<=h2.data:
        head=h1
        h1=h1.next
    else:
        head=h2
        h2=h2.next
    curr=head
    while (h1 and h2):
        if h1.data<=h2.data:
            curr.next=h1
            h1=h1.next
        else:
            curr.next=h2
            h2=h2.next
        curr=curr.next
    if h1==None:
        curr.next=h2
    else:
        curr.next=h1
    return head
  
p=ll2()
print("Enter elements of first list")
p.ins_end()
q=ll2()
print("Enter elements of second list")
q.ins_end()
z=merge(p.start,q.start)
traverse(z)
```

### Bubble Sort in LL

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
class ll2:
    def __init__(self):
        self.start=None
    def ins_end(self):
        print("Enter no of elements:")
        for i in range(int(input())):
            print("Enter element")
            n=node(int(input()))
            n.next=None
            if self.start==None:
                self.start=n
            else:
                ptr=self.start
                while(ptr.next!=None):
                    ptr=ptr.next
                ptr.next=n
    def bubbleSort(self):
        if self.start==None or self.start.next==None:
            return self.start
        swap=True
        while swap:
            swap=False
            prev=None
            curr=self.start
            nex=curr.next
            while nex:
                if curr.data>nex.data:
                    swap=True
                    if prev:
                        prev.next=nex
                    else:
                        self.start=nex
                    curr.next=nex.next
                    nex.next=curr
                    prev=nex
                else:
                    prev=curr
                curr=prev.next
                nex=curr.next
    
    def traverse(self):
        current=self.start
        while(current!=None):
            print(current.data)
            current=current.next
p=ll2()
p.ins_end()
p.bubbleSort()
p.traverse()
```
